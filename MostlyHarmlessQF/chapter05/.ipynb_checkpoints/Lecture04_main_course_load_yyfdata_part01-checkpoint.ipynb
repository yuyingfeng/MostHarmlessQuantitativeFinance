```python
def estimated_autocorrelation(x):
    n = np.size(x)
    var_x = x.var()
    x = x-x.mean()
    r = np.correlate(x, x, mode = 'full')[-n:]
    #assert N.allclose(r, N.array([(x[:n-k]*x[-(n-k):]).sum() for k in range(n)]))
    result = r/(var_x*(np.arange(n, 0, -1)))
    return result

yy = estimated_autocorrelation(xx)
```

```python
def gjr_garch_likelihood(parameters, data, sigma2, out = None):
    mu = parameters[0]
    omega = parameters[1]
    alpha = parameters[2]
    gamma = parameters[3]
    beta = parameters[4]
    
    T = np.size(data,0)
    #print('What is T=',T)
    eps = data - mu
    # Data and Sigma2 are T by 1 vectors
    for t in range(1,T):
        sigma2[t]=(omega+alpha*eps[t-1]**2+gamma*eps[t-1]**2 * (eps[t-1]<0)+beta*sigma2[t-1])
        
    logliks = 0.5*(np.log(2*np.pi)+np.log(sigma2)+eps**2/sigma2)
    loglik = np.sum(logliks)
    
    if out is None:
        return loglik
    else:
        return loglik, logliks, np.copy(sigma2)
```

```python
startingVals = np.array([dfrets.SZINDEX.mean(),
                       dfrets.SZINDEX.var()*.01,
                       .03,.09,.90])
T=len(dfrets.SZINDEX);
sigma2 = np.ones(T)*(dfrets.SZINDEX).var()
#estimates
#np.size(sigma2)
loglik0, logliks, sigma2final = gjr_garch_likelihood(startingVals,dfrets.SZINDEX, sigma2, out=True)

vol = pd.DataFrame(np.sqrt(252*sigma2),index=dfrets.index,columns=['volatility'])
%matplotlib inline
vol.plot(grid='on',color = '#CC0066',title='SZ volatility with GJR')

vol.describe()
vol
```

```python
from numpy.linalg import inv
from scipy.optimize import fmin_slsqp

def gjr_constraint(parameters, data, sigma2, out=None):
    alpha = parameters[2]
    gamma = parameters[3]
    beta = parameters[4]
    return np.array([1-alpha-gamma/2-beta])

finfo=np.finfo(np.float64)

bounds  =[(-10*dfrets.SZINDEX.mean(),10*dfrets.SZINDEX.mean()),
         (finfo.eps,2*dfrets.SZINDEX.var()),
         (0.0,1.0),(0.0,1.0),(0.0,1.0)]
args = (np.asarray(dfrets.SZINDEX),sigma2)
estimates = fmin_slsqp(gjr_garch_likelihood, startingVals, f_ieqcons=gjr_constraint, bounds=bounds, args =args)
print('Initial Values=',startingVals)
print('Estimated Values=',estimates)
```

```python
loglik, logliks, sigma2final = gjr_garch_likelihood(estimates,dfrets.SZINDEX, sigma2, out=True)

vol2 = pd.DataFrame(np.sqrt(252*sigma2),index=dfrets.index,columns=['volatility'])
%matplotlib inline
vol2.plot(grid='on',color = '#32CD32',title='SZ volatility with estimated GJR')
print('initial loglik=',loglik0,'ested loglik=',loglik)
vol2.describe()

vol.merge(vol2)
```
